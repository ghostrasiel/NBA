```python
import sqlite3
conn = sqlite3.connect("./basketball.sqlite")
result = conn.execute("select namePlayerBREF , urlPlayerBioBREF from Player_Bios")
sqldata = result.fetchall()
data =[]
a = 0
for d in sqldata:
    a +=1
    json={'name':d[0] , 'url':d[1]}
    data.append(json)
print(a)
conn.close()
```

```python
import pandas as pd
df = pd.DataFrame(data)
df = df[~df.duplicated()]
df.head(5)
```

```python
import requests
from bs4 import BeautifulSoup
import time
import threading


header ={'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.164 Safari/537.36'}
sql = f"insert into player_data values({'?,'*30}?)"

def nba_requests(urls):
    conn = sqlite3.connect("./basketball.sqlite")
    global header
    global sql
    b = 0
    for url in urls:
        try:
            res = requests.get(url = url , headers = header)
            soup = BeautifulSoup(res.text , 'html.parser')
            tables = soup.select('table#per_game tbody tr')
            name = soup.select('div#meta div h1 span')[0].text
            a = 0
            for t in tables:
                a +=1
                t_data = list()
                tds = t.select('td')
                th = t.select('th')[0].text
                t_data.append(name)
                t_data.append(th)
                for td in tds:
                    t_data.append(td.text)
                conn.execute(sql , tuple(t_data))

            if b%20 == 0 :
                conn.commit()

            print('ok' , name , f'共擁有{a}個賽季')
            time.sleep(2)
        except:
            pass

    conn.commit()
    conn.close()

start = time.time()

urls = df['url'].values.tolist()
urls_1 = [urls[0:1336]]
urls_2 = [urls[1336:2672]]
urls_3 = [urls[2672:4010]]
t_list = list() #定議線程

t1 = threading.Thread(target=nba_requests, args=(urls_1))
t_list.append(t1)
t2 = threading.Thread(target=nba_requests, args=(urls_2))
t_list.append(t2)
t3 = threading.Thread(target=nba_requests, args=(urls_3))
t_list.append(t3)

for t in t_list:
    t.start()

for t in t_list:
    t.join()    

 
print('ok')
end = time.time()
print('花費時間:',end - start)

    
```

```python
import threading
import time

def main(url, num):
    print('開始執行', url)
    time.sleep(2)
    print('結束', num)

url_list1 = ['www.yahoo.com.tw, www.google.com' , 'xxxx' , 'dddd']
url_list2 = ['www.yahoo.com.tw, www.google.com']
url_list3 = ['www.yahoo.com.tw, www.google.com']

# 定義線程
t_list = []

t1 = threading.Thread(target=main, args=(url_list1, 1))
t_list.append(t1)
t2 = threading.Thread(target=main, args=(url_list2, 2))
t_list.append(t2)
t3 = threading.Thread(target=main, args=(url_list3, 3))
t_list.append(t3)

# 開始工作
for t in t_list:
    t.start()

# 調整多程順序
for t in t_list:
    t.join()
```
